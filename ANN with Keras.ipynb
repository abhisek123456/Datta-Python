```python
import pandas as pd
import os
os.chdir("/home/abhisek/Studies/Python program/Data Kaggle/Practice/Data")
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df.head(10)
```

```python
lst_geo=df["Geography"].value_counts().head(2).index
lst_geo=list(lst_geo)
lst_geo
```

```python
import numpy as np
for geo in lst_geo:
    df[geo]=np.where(df["Geography"]==geo,1,0)
```

```python
df.head(10)
```

```python
lst_gen=df["Gender"].value_counts().head(1).index
lst_gen=list(lst_gen)
lst_gen
```

```python
for gen in lst_gen:
    df[gen]=np.where(df["Gender"]==gen,1,0)
```

```python
df.head(10)
```

```python
# Independent variables
X=df.iloc[:,[3,6,7,8,9,10,11,12,14,15,16]]
X.head()
```

```python
# dependent variable
y=df.iloc[:,[13]]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=0)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#Initializating ANN 
ANN=Sequential()
```

```python
#Imput and First Layer 
ANN.add(Dense(units = 9, kernel_initializer = 'he_uniform',activation='relu',input_dim = 11))
```

```python
ANN.add(Dense(units = 5, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
ANN.add(Dense(units = 5, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
ANN.add(Dense(units = 1, kernel_initializer = 'glorot_uniform',activation='sigmoid'))
```

```python
ANN.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
ANN.summary()
```

```python
ANN.fit(X_train, y_train,validation_split=0.33, batch_size = 10,epochs= 100)
```

```python
y_pred=ANN.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn import metrics
cnf=metrics.confusion_matrix(y_test,y_pred)
cnf
```

```python
print(metrics.classification_report(y_test,y_pred))
```
