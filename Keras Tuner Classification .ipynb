```python
import pandas as pd
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers 
from kerastuner import RandomSearch
import os
os.chdir("/home/abhisek/Studies/Python program/Data Kaggle/Practice/Data")
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
X=df.iloc[:,[3,6,7,8,9,10,11,12]]
X.head()
```

```python
y=df.iloc[:,[13]]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
def built_model(hp):
    model=keras.Sequential()
    for i in range (hp.Int("num_layer",2,20)):
        model.add(layers.Dense(units=hp.Int("units_"+ str(i),min_value=32,max_value=512,step=32),
                               activation="relu"))
        model.add(layers.Dense(1,activation="sigmoid"))
        model.compile(optimizer=keras.optimizers.Adam(hp.Choice("learning rate",[0.01,0.001,0.0001])),
                      loss="binary_crossentropy",
                      metrics=["accuracy"])
    return model           
```

```python
tuner=RandomSearch(built_model,
                  objective="val_accuracy",
                      max_trials=3,
                  directory="project1")
```

```python
tuner.search_space_summary()
```

```python
tuner.search(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```
